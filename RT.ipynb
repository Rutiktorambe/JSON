# Operators

Operators perform operations on variables.

## Arithmetic Operators

| Operator | Meaning        |
| -------- | -------------- |
| +        | Addition       |
| -        | Subtraction    |
| \*       | Multiplication |
| /        | Division       |
| \*\*     | Power          |

````python
Number01 = 64
Number02 = 6

Sum = Number01 + Number02
print("Sum:", Sum)

Difference = Number01 - Number02
print("Difference:", Difference)

Product = Number01 * Number02
print("Product:", Product)

Division= Number01 / Number02
print("Division:", Division)

Square_of_Number01 = Number01 ** 2
print("Square of Number01:", Square_of_Number01)
````

# 5. Dictionaries and Dictionary Manipulation

## Definition

A **dictionary** in Python is a data structure used to store information in **key–value pairs**.

It works like an **office information manager** who stores organized records so information can be quickly retrieved.

Structure:

```
key : value
```

Example:

```python
employee_info = {
    "name": "Rahul",
    "department": "IT",
    "experience": 3,
    "coffee_per_day": 2,
    "skills": ["Python", "SQL"],
    "manager": {
        "name": "Sneha",
        "department": "Engineering"
    }
}
```

---

# Dictionary as an Information Manager

Imagine an **office information manager** keeping records of employees.

Instead of searching through piles of papers, the manager keeps organized data like:

```
Name → Rahul
Department → IT
Experience → 3 years
Coffee Per Day → 2
Manager → Sneha
```

Whenever someone asks for information, the manager quickly finds it using the **key**.

Python dictionaries behave the same way.

---

# Important Notes

• Dictionaries use **curly brackets `{}`**
• Data is stored as **key : value** pairs
• Keys must be **unique**
• Values can be **any data type**
• Dictionaries are **mutable (changeable)**

---

# Our Example Dictionary

We will use **one dictionary throughout the examples**.

```python
employee_info = {
    "name": "Rahul",
    "department": "IT",
    "experience": 3,
    "coffee_per_day": 2,
    "skills": ["Python", "SQL"],
    "manager": {
        "name": "Sneha",
        "department": "Engineering"
    }
}
```

This dictionary contains:

* Strings
* Numbers
* List
* Nested dictionary

---

# Accessing Dictionary Values

```python
print(employee_info["name"])
print(employee_info["department"])
```

Output

```
Rahul
IT
```

---

# Accessing Nested Dictionary

```python
print(employee_info["manager"]["name"])
```

Output

```
Sneha
```

Explanation:

The dictionary **manager** contains another dictionary.

---

# Adding New Information

Suppose the information manager adds Rahul's **salary**.

```python
employee_info["salary"] = 60000
```

Dictionary becomes:

```python
{
"name": "Rahul",
"department": "IT",
"experience": 3,
"coffee_per_day": 2,
"skills": ["Python", "SQL"],
"manager": {"name": "Sneha", "department": "Engineering"},
"salary": 60000
}
```

---

# Updating Information

Rahul got a raise.

```python
employee_info["salary"] = 65000
```

Now the information manager updates the record.

---

# Removing Information

Suppose coffee tracking is removed from the system.

```python
del employee_info["coffee_per_day"]
```

Another way using `pop()`:

```python
employee_info.pop("experience")
```

---

# Using get() (Safe Access)

Sometimes the key may not exist.

```python
print(employee_info.get("location"))
```

Output

```
None
```

This avoids errors.

---

# Getting All Keys

```python
print(employee_info.keys())
```

Output

```
dict_keys(['name','department','skills','manager','salary'])
```

---

# Getting All Values

```python
print(employee_info.values())
```

---

# Getting Key-Value Pairs

```python
print(employee_info.items())
```

---

# Looping Through the Dictionary

The office information manager reads every record.

```python
for key, value in employee_info.items():
    print(key, ":", value)
```

Example output:

```
name : Rahul
department : IT
skills : ['Python', 'SQL']
manager : {'name': 'Sneha', 'department': 'Engineering'}
salary : 65000
```

---

# Accessing Nested Manager Info

```python
print(employee_info["manager"]["department"])
```

Output

```
Engineering
```

Meaning:

Rahul's manager Sneha works in the **Engineering department**.
