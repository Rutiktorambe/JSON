## Selection & Filtering


### df['col']
- What: Select column
- When: Access data
- Returns: Series

### df.loc[]
- What: Label-based selection
- When: Index labels
- Returns: Series/DataFrame

### df.iloc[]
- What: Position-based selection
- When: Index numbers
- Returns: Series/DataFrame


```
df['Sales']
df.loc[0]
df.iloc[:2]
df[df['Sales']>500]
```

## Operations


### Column operations
- What: Create/modify columns
- When: Feature engineering
- Returns: Updated DataFrame


```
df['Margin'] = df['Profit']/df['Sales']
```

### 📝 Practice
Create double sales column

```
df['DoubleSales']=df['Sales']*2
```

## Grouping


### df.groupby()
- What: Split data into groups
- When: Aggregation/reporting
- Returns: GroupBy object


```
df.groupby('Region')['Sales'].sum()
```

### 📝 Practice
Average profit by category

```
df.groupby('Category')['Profit'].mean()
```

## Merge


### pd.merge()
- What: Combine datasets
- When: SQL-like join
- Returns: DataFrame


```
pd.merge(df, df, on='OrderID')
```

## Pivot


### pd.pivot_table()
- What: Reshape data
- When: Reporting
- Returns: DataFrame


```
pd.pivot_table(df, values='Sales', index='Region', columns='Category')
```

### df.fillna()
- What: Replace NaN

- Returns: DataFrame

### df.dropna()
- What: Remove NaN rows

```
df.fillna(0)
df.dropna()
```

## Date & Time


### pd.to_datetime()
- What: Convert to datetime
- When: Date processing
- Returns: Series

### .dt.year
- What: Extract year
- When: Analysis
- Returns: Series


```
df['Date']=pd.to_datetime(df['Date'])
df['Year']=df['Date'].dt.year
```
