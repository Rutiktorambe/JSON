```
df = pd.read_csv('/mnt/data/training_sales.csv')
df.head()
df.info()
df.describe()
```

## Selection & Slicing

loc, iloc, filtering

```
df['Sales']
df.loc[0]
df.iloc[:2]
df[df['Sales']>500]
```

```
df[df['Profit']>100]
```

Create new columns

```
df['Margin'] = df['Profit']/df['Sales']
```

```
df.rename(columns={'Sales':'TotalSales'})
```

```
df.fillna(0)
df.dropna()
```

```

df2 = df.copy()
pd.merge(df, df2, on='OrderID')
pd.concat([df, df2])
```
